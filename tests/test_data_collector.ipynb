```python
import mesa
```

```python
class MyModel(mesa.Model):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.x = 0
        self.datacollector = mesa.DataCollector(
            model_reporters={
                "x": "x",
            }
        )
    
    def step(self):
        self.datacollector.collect(model=self)
        self.x += 1

model = MyModel()
model.step()
model.step()
model.step()
model.step()
```
